```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

train_accuracy = df['train_accuracy']
labels = df['labels']

colors = plt.cm.YlOrRd(train_accuracy)

fig, ax = plt.subplots(figsize=(15, 10))

bars = ax.bar(x=labels, height=train_accuracy, color=colors)

ax.set_title('train accuracy')
ax.set_xlabel('optymalizacja danego parametru / struktury sieci')
ax.set_ylabel('accuracy')
ax.set_xticklabels(labels, rotation=90)

sm = plt.cm.ScalarMappable(cmap='YlOrRd')
sm.set_array([])
fig.colorbar(sm, ax=ax, label='train accuracy')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

val_accuracy = df['val_accuracy']
labels = df['labels']

colors = plt.cm.YlOrRd(val_accuracy)

fig, ax = plt.subplots(figsize=(15, 10))

bars = ax.bar(x=labels, height=val_accuracy, color=colors)

ax.set_title('validation accuracy')
ax.set_xlabel('optymalizacja danego parametru / struktury sieci')
ax.set_ylabel('accuracy')
ax.set_xticklabels(labels, rotation=90)

sm = plt.cm.ScalarMappable(cmap='YlOrRd')
sm.set_array([])
fig.colorbar(sm, ax=ax, label='validation accuracy')

plt.tight_layout()
plt.show()
```

```python
time = df['time']
labels = df['labels']

norm = plt.Normalize(time.min(), time.max())
colors = plt.cm.YlOrRd(norm(time))

fig, ax = plt.subplots(figsize=(15, 10))

bars = ax.bar(x=labels, height=time, color=colors)

ax.set_title('ile czasu zajelo wytrenowanie w minutach')
ax.set_xlabel('optymalizacja danego parametru / struktury sieci')
ax.set_ylabel('czas')
ax.set_xticklabels(labels, rotation=90)

sm = plt.cm.ScalarMappable(cmap='YlOrRd', norm=norm)
sm.set_array([])
fig.colorbar(sm, ax=ax, label='time')

plt.tight_layout()
plt.show()
```
